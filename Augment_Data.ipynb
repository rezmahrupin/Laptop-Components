```python
import os
from keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
from shutil import copy2

# Specify your image directory
image_folder = r'/content/drive/MyDrive/Capstone_NEW_DATASET_140/1. Battery'

# Specify the directory to save augmented and original images
output_folder = r'/content/drive/MyDrive/Augmented/1. Battery'
os.makedirs(output_folder, exist_ok=True)  # Create the directory if it doesn't exist

# Create an ImageDataGenerator instance with specified augmentation parameters
datagen = ImageDataGenerator(
    rotation_range=45,
    zoom_range=0.2,
    horizontal_flip=True,
)

# Get a list of all image files in the directory
image_files = [os.path.join(image_folder, f) for f in os.listdir(image_folder) if f.endswith(('.jpg', '.jpeg', '.png'))]

# Check if there are at least 2 images
if len(image_files) < 2:
    print("Error: Not enough images in the directory.")
else:
    # Process only the first 2 images
    for img_path in image_files[:140]:
        # Load the image
        img = load_img(img_path)
        x = img_to_array(img)
        x = x.reshape((1,) + x.shape)

        # Copy original image to the new folder
        original_img_name = os.path.basename(img_path)
        copy2(img_path, os.path.join(output_folder, original_img_name))

        # Generate 3 augmented images and save to the new directory
        for i in range(3):  # Generate 3 augmented copies for each image
            batch = next(datagen.flow(x, batch_size=1, save_to_dir=output_folder, save_prefix=f'aug_{i}', save_format='jpg'))

print("Data augmentation complete.")
```

```python
import os
from keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
from shutil import copy2
from PIL import ImageFile

# Allow loading of truncated images
ImageFile.LOAD_TRUNCATED_IMAGES = True

# Specify your image directory
image_folder = r'/content/drive/MyDrive/Capstone_NEW_DATASET_140/23. RAM_Cover'

# Specify the directory to save augmented and original images
output_folder = r'/content/drive/MyDrive/Augmented/23. RAM_Cover'
os.makedirs(output_folder, exist_ok=True)  # Create the directory if it doesn't exist

# Create an ImageDataGenerator instance with specified augmentation parameters
datagen = ImageDataGenerator(
    rotation_range=45,
    zoom_range=0.2,
    horizontal_flip=True,
)

# Get a list of all image files in the directory
image_files = [os.path.join(image_folder, f) for f in os.listdir(image_folder) if f.endswith(('.jpg', '.jpeg', '.png', '.JPG', '.webp'))]

# Check if there are at least 2 images
if len(image_files) < 139:
    print("Error: Not enough images in the directory.")
else:
    # Process only the first 2 images
    for img_path in image_files[:139]:
        try:
            # Load the image
            img = load_img(img_path)
            x = img_to_array(img)
            x = x.reshape((1,) + x.shape)

            # Copy original image to the new folder
            original_img_name = os.path.basename(img_path)
            copy2(img_path, os.path.join(output_folder, original_img_name))

            # Generate 3 augmented images and save to the new directory
            for i in range(3):  # Generate 3 augmented copies for each image
                batch = next(datagen.flow(x, batch_size=1, save_to_dir=output_folder, save_prefix=f'aug_{i}', save_format='jpg'))
        except OSError as e:
            print(f"Error processing {img_path}: {e}")

print("Data augmentation complete.")
```
